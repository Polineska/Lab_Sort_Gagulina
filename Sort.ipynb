```python
import matplotlib.pyplot as plt
data=[10,100,1000, 10000, 100000]
time=[4, 1, 5,  39, 425 ]
plt.xlabel('t')
plt.ylabel('Size of array')
plt.title('Sorting time Quick Sort')
plt.plot(data,time)
```

```python
import matplotlib.pyplot as plt
data=[10,100,1000, 10000, 100000]
time=[0, 0, 3,  345, 52341 ]
plt.xlabel('t')
plt.ylabel('Size of array')
plt.title('Sorting time Bubble Sort')
plt.plot(data,time)
plt.plot(data,time)
```

```python
import matplotlib.pyplot as plt
data=[10,100,1000, 10000, 100000]
time=[0, 0, 0, 0, 3 ]
plt.xlabel('t')
plt.ylabel('Size of array')
plt.title('Sorting time Insertion Sort')
plt.plot(data,time)
plt.plot(data,time)
```

```python
plt.plot([4, 1, 5,  39, 425], [10,100,1000, 10000, 100000], color='blue', label = 'Quick Sort')
plt.plot([0, 0, 3,  345, 52341], [10,100,1000, 10000, 100000], color='green', label = 'Bubble Sort')
plt.plot([0, 0, 0, 0, 3 ], [10,100,1000, 10000, 100000], color='red', label = 'Insertion Sort')
plt.xlabel('t, мс')
plt.ylabel('Size of array')
plt.title('Sorting time')
plt.legend()
plt.show()
```
